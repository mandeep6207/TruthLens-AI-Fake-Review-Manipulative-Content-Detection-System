## Architecture and Workflow

TruthLens AI uses a modular workflow:

1. Generate synthetic review text with noise and realism.
2. Clean and normalize the text for NLP modeling.
3. Engineer review behavior features such as length and sentiment.
4. Train TF-IDF-based models with regularization.
5. Compare classifiers and save evaluation artifacts.

This structure keeps the project easy to explain to reviewers and easy to extend later.
